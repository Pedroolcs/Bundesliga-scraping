```python
years = list(range(2024,2022, -1))
```

```python
years
```

```python
all_matches = []
```

```python
standings_url = 'https://fbref.com/pt/comps/20/Bundesliga-Estatisticas'
```

```python
for year in years:
    response = requests.get(standings_url)
    soup = BeautifulSoup(response.text)
    standings_table = soup.select('table.stats_table')[0]

    links = [l.get('href') for l in standings_table.find_all('a')]
    links = [l for l in links if '/equipes/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]

    previous_season = soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com/{previous_season}"
    
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Estatisticas","").replace("-", " ")
        response = requests.get(team_url)
        matches = pd.read_html(StringIO(response.text), match='Resultados e Calendários')[0]
        soup = BeautifulSoup(response.text)
        links = [l.get('href') for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        response = requests.get(f"https://fbref.com{links[0]}")
        shooting = pd.read_html(StringIO(response.text), match="Chutes")[0]
        shooting.columns = shooting.columns.droplevel()

        try:
            team_data = matches.merge(shooting[["Data", "TC", "CaG", "Dist", "FK"]], on="Data")
        except ValueError:
            continue

        team_data = team_data[team_data["Camp."] == "Bundesliga"]
        team_data["Temporada"] = year
        team_data["Equipe"] = team_name
        all_matches.append(team_data)
        time.sleep(1)
```

```python
len(all_matches)
```

```python
match_df = pd.concat(all_matches)
```

```python
match_df.columns = [c.lower() for c in match_df.columns]
```

```python
match_df.to_csv("matches.csv")
```
